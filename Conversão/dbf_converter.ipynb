```python
from dbfread import DBF
import pandas as pd


dbf = DBF('ORDEM-SERV.DBF')
dataResult = pd.DataFrame(iter(dbf))

display(dataResult)
```

```python

tabela = dataResult[['NR_ORDEM','NRSERIE', 'DATAFECHA', 'CODCLI', 'NOMECLIENT']]
df = tabela[tabela['CODCLI'].str.contains('147')]
#filtro_data = filtro_cliente[(filtro_cliente['DATAFECHA'] >= '2024-01-01') & (filtro_cliente['DATAFECHA'] <= '2024-01-30')]

display(df)
```

```python
df['DATAFECHA'] = pd.to_datetime(df['DATAFECHA'], errors='coerce')

datai = input()
dataf = input()

data_inicial = pd.to_datetime(datai)
data_final = pd.to_datetime(dataf)

df_filtrado = df[(df['DATAFECHA'] >= data_inicial) & (df['DATAFECHA'] <= data_final)]

display(df_filtrado)
```
